```python
import pandas as pd
import os
```

```python
reco_file_dir = "../Final_Recommendation_Letters/"
candidates = pd.read_csv("../Final_Resume_Texts/resume_with_actual_roles.csv")
# add a lor field to the candidates dataframe
candidates["lor_data"] = ""
# add a lor_from field to the candidates dataframe with empty list
candidates["lor_from"] = [[] for i in range(len(candidates))]
for i in range(len(candidates)):
    
    reco_data = ""
    id = candidates.loc[i, "id"]
    # in the folder, sub folder of the form Recommendation_Letters_of_ID_1
    # files are of the form Recommendation_From_ID_2.txt
    reco_file_sub_dir = reco_file_dir + "Recommendation_Letters_of_ID_" + str(id) + "/"
    # if the folder does not exist, skip
    if not os.path.exists(reco_file_sub_dir):
        continue
    for letter in os.listdir(reco_file_sub_dir):
        with open(reco_file_sub_dir + letter, "r", encoding="windows-1250") as f:
            reco_data += f.read()
        candidates.loc[i, "lor_from"].append(int(letter.split("_")[-1].split(".")[0]))
    candidates.loc[i, "lor_data"] = reco_data

candidates.head()
```

```python
candidates.to_csv("../Final_Resume_Texts/resume_with_actual_roles_and_lor.csv", index=False)
```
