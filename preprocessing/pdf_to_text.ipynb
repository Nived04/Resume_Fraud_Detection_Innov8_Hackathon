```python
import textract
import pandas as pd
import json
import os
from pypdf import PdfReader


DIR = "../Final_Resumes/"
OUT_DIR = "../Final_Resume_Texts/"
```

```python
texts = {}
from tqdm import tqdm
# file name is Resume_of_ID_num.pdf

# make the dir if it doesn't exist
if not os.path.exists(OUT_DIR):
    os.makedirs(OUT_DIR)

cnt = 0
for file in tqdm(os.listdir(DIR)):
    cnt += 1
    try:
        if file.endswith(".pdf"):
            text = "" 
            reader = PdfReader(DIR + file)
            for page in reader.pages:
                text += page.extract_text()
            id = file.split('_')[-1].split('.')[0]
            texts[id] = text
    except Exception as e:
        print(e)
        print(file)
        continue
    # if (cnt == 5):
    #     break
    
# with open(OUT_DIR + 'resume_texts.json', 'w') as f:
#     json.dump(texts, f)
```

```python
# get a list of resume titles. It is the first line in each resume with length greater than 4. Store the counts
titles = {}
for key in texts:
    lines = texts[key].split('\n')
    for line in lines:
        if(len(line) >= 5):
            title = line.lower()
            titles[key] = title
            break
        
print(titles)
```

```python
# create a dataframe with id, title, text
df = pd.DataFrame(columns=['id', 'title', 'text'])
for key in texts:
    new_row = pd.DataFrame({'id': key, 'title': titles[key], 'text': texts[key]}, index=[0])
    df = pd.concat([df, new_row], ignore_index=True)
    
df.to_csv(OUT_DIR + 'resume_texts.csv', index=False)
```
